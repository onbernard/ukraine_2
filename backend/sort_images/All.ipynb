```python
from pathlib import Path
from collections import defaultdict
import polars as pl
from index import index
root_path = Path(r"/mnt/z/DATASETS/ukraine/russia/russia")
df = pl.read_csv(r"/mnt/z/DATASETS/ukraine/losses_russia.csv")
```

```python
for p in root_path.rglob("*.*"):
    matches = [sub.get(' '.join(p.stem.split("_")[:-1])) for sub in index.values()]
    matches = [match for match in matches if match]
    if len(matches)!=1:
        print(p)
        print(matches)
```

```python
images = defaultdict(list)
for p in root_path.rglob("*.*"):
    name = ' '.join(p.stem.split("_")[:-1])
    matches = [sub.get(' '.join(p.stem.split("_")[:-1])) for sub in index.values()]
    match = [match for match in matches if match][0]
    images[match].append(p)
images
```

```python
img_col = []
for r in df.rows():
    imgs = ['/'+'/'.join(p.parts[-3:]) for p in images[r[1]]]
    img_col.append(imgs)
img_col
```

```python
df_with_images = df.with_columns(pl.Series(name="images", values=[':'.join(paths) for paths in img_col]))
```

```python
df_with_images.write_csv("losses_russia_with_images.csv")
```
